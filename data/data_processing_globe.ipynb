```python
# Make the year before Christ negative
all['Year'] = all['Year'].apply(lambda x: "-" + x.split(' ')[0] if 'BC' in x else x)
all['Year'] = pd.to_numeric(all['Year'], errors='coerce')
# Drop rows with missing years
all.dropna(subset=['Year'], inplace=True)
```

```python
#copy the data
coord = all.copy()
```

```python
# Initialize the geocoder
geolocator = Nominatim(user_agent="geoapiRun1")

# To prevent hitting the service too hard, wrap the geolocator with a rate limiter
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=2)

# Function to apply to each row in the DataFrame
def geocode_place(row):
    # Combine the place name and country into one search query for more accurate results
    query = f"{row['Place Name']}, {row['Country']}"
    try:
        # Attempt to geocode using the combined query
        location = geocode(query)
        if location:
            return pd.Series([location.latitude, location.longitude])
        else:
            return pd.Series([None, None])
    except:
        return pd.Series([None, None])

# Apply the function to the DataFrame
coord[['latitude', 'longitude']] = coord.apply(geocode_place, axis=1)
```

```python
# Initialize the geocoder for run two (on the data that failed in run one)
geolocator = Nominatim(user_agent="geoapiRun2")

# To prevent hitting the service too hard, wrap the geolocator with a rate limiter
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=2)

# Function to apply to each rowin the DataFrame
def geocode_place(row):
    # If the latitude is already known, return it
    if pd.notnull(row['latitude']):
        return pd.Series([row['latitude'], row['longitude']])
    # Combine the place name and country into one search query for more accurate results
    query = f"{row['PlaceName']}, {row['Country']}"
    try:
        # Attempt to geocode using the combined query
        location = geocode(query)
        if location:
            return pd.Series([location.latitude, location.longitude])
        else:
            return pd.Series([None, None])
    except:
        return pd.Series([None, None])

# Apply the function to the DataFrame
coord[['latitude', 'longitude']] = coord.apply(geocode_place, axis=1)
```

```python
#Mapping to name of the columns to be used in javascript
rename_mapping = {
    'Name of Incident': 'NameOfIncident',
    'Type of Event': 'TypeOfEvent',
    'Place Name': 'PlaceName',
    'Affected Population': 'AffectedPopulation',
    'Important Person/Group Responsible': 'ImportantPersonGroupResponsible',
}

coord = coord.rename(columns=rename_mapping)
```

```python
countries_freq = all['Country'].value_counts().reset_index()
countries_freq.columns = ['Country', 'Frequency']
countries_freq.head()
```

```python
# Load a country geojson file
gdf = gpd.read_file("ne_110m_admin_0_countries.geojson")
```

```python
countries_freq['Country'] = countries_freq['Country'].str.title()
gdf['ADMIN'] = gdf['ADMIN'].str.title()
```

```python
# Merge the geodataframe with the country frequencies
merged_gdf = gdf.merge(countries_freq, left_on='ADMIN', right_on='Country', how='left')
```

```python
merged_gdf['Frequency'] = merged_gdf['Frequency'].fillna(0)
```

```python
# save the updated GeoJSON file
merged_gdf.to_file("countries.geojson", driver='GeoJSON')
```
